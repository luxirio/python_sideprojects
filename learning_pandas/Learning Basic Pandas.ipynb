### Loading data into Pandas

```python
import pandas as pd #importing the library

df = pd.read_csv("pokemon_data.csv") # reading a csv file
#df_txt = pd.read_csv("pokemon_data.txt", delimiter = "\t") # reading txt files, use the delimiter "\t"
#df_excel = pd.read_excel("pokemon_data.xlsx", sheet_name = 0) # reading excel file

df.head(10)
```

```python
## reading column names
#print(df.columns)

## reading each column
# df[["Name", "Type 1"]] #reading through column names

## reading each row
## reading row by index index localization or iloc
# df.iloc[0:4]

# iterating through rows:
#for index, row in df.iterrows():
    #print(index, row["Name"]) #only the name of each row

df.loc[df["Type 1"] == "Grass"] #filtering/using the query type
```

### Sorting Values in our dataset

```python
#df.describe() #describing data using

df.sort_values(["Type 1", "HP"],  ascending = [1,0]) #we can have different ascending, descending using multiple columns
```

### Making changes to the data

```python
#creating a new column making operations in another columns
#df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]

#df = df.drop(columns = ["Total"]) #droping a column

df["Total"] = df.iloc[:,4:10].sum(axis = 1) #axis = 1 summing horizontally axis = 0 summing vertically
#in this case if axis = 0, total = NaN

#another way to make changes to the data is using columns list
cols = list(df.columns)
#to re-arrange the columns, one way of doing it:
df = df[cols[0:4] + cols[-1:] + cols[4:12]]
#another way to re-indexing the columns order is:
#df = df[["#", "Name", "Type 1", "Type 2", "Total"]]

df.head(3)
```

### Writing files

```python
df.to_csv("modified.csv", index = False) #to save a csv file
df.to_excel("modified.xlsx", index = False) #to save an excel file

# df.to_csv("modified.txt", index = False, delimiter ="\t") #save a txt file
```

### Filtering data

```python
#df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison") & (df["HP"] > 70)].reset_index(drop = True) #different conditions BETWEEN PARENTHESIS
## it is also important to reset the INDEX AFTER FILTERING

import re

df.loc[~df["Name"].str.contains("^pi[a-z]*", regex = True, flags = re.I)] #the ~ is the NOT ^pi[a-z] starts with pi or Pi and any letter a through z after
```

### Conditional Changes

```python
#basically it
df.loc[df["Total"] > 500, ["Legendary", "Generation"]] = [True, 10] #multiple condition assigments
#in this case if column total > 500, legendary = true and generation = 10
```

### Aggregate Statistics using Group_by

```python
df = pd.read_csv("modified.csv")
df.head(3)
```
